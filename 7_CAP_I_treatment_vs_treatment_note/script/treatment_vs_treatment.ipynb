## Treatment vs Treatment Heatmap
calculate correlation coefficient between each treatments.

```python
import pandas as pd
import numpy as np

# log_matrix_diff_expression.csv from AtacR.Rmd
log_matrix_data = pd.read_csv("../data/log_matrix_diff_expression.csv", index_col=0)
log_matrix_data = log_matrix_data.drop("AT4G28410_RSA1", axis=0)

# calculate coefficient
coef_values = pd.DataFrame(np.corrcoef(log_matrix_data.T))
coef_values.columns = log_matrix_data.columns
coef_values.index = log_matrix_data.columns
coef_values
```

## Make heatmap

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline
%config InlineBackend.figure_formats = {'png', 'retina'}

heat_map_values = coef_values.loc[coef_values.index != "gdna", coef_values.columns != "gdna"]

heatmap = sns.clustermap(heat_map_values, cmap="BuGn", annot=False)
plt.show()
```

## Add colors for mutants/treatments (circle version)

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%config InlineBackend.figure_formats = {'png', 'retina'}

y_colors = {
    "eds1":(0.9,0.6,0),
    "gh":(0.35,0.7,0.9),
    "myc234":(0,0.6,0.5),
    "wt":(0,0,0),
    "r1ab":(0.95,0.9,0.25),
    "tplr14":(0,0.45,0.7),
    "peds":(0.8,0.4,0),
    "sid2":(0.8,0.6,0.7)
}

x_colors = {
    "mk":(0,0,0),
    "a2":(0.8,0.4,0),
    "a4":(0.8,0.4,0),
    "kv":(0.35,0.7,0.9)
}

heat_map_values = coef_values.loc[coef_values.index != "gdna", coef_values.columns != "gdna"]

heatmap = sns.clustermap(heat_map_values, cmap="BuGn", annot=False) #vmin=0.6)
ytl = heatmap.ax_heatmap.get_yticklabels()
for each_y in ytl:
    texts = each_y.get_text()
    each_y.set_color(y_colors[texts[:texts.find("_")]])
    
xtl = heatmap.ax_heatmap.get_xticklabels()
for each_x in xtl:
    texts = each_x.get_text()
    each_x.set_color(x_colors[texts[texts.find("_")+1:]])
labels = ["â—" for x in range(heat_map_values.shape[0])]
heatmap.ax_heatmap.set_xticklabels(labels, fontsize = 20)
heatmap.ax_heatmap.set_yticklabels(labels, fontsize = 20)
plt.show()
```
