# Make additional linear plot related to Time Series heatmap

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%config InlineBackend.figure_formats = {'png', 'retina'}

# read dataset. each z_score dataset from Time_series_TPM.Rmd
control_Pop2 = pd.read_csv("../data/control_z_score_Pop2.csv", index_col=0)
target_Pop2 = pd.read_csv("../data/target_z_score_Pop2.csv", index_col=0)
control_C321A = pd.read_csv("../data/control_z_score_C321A.csv", index_col=0)
target_C321A = pd.read_csv("../data/target_z_score_C321A.csv", index_col=0)
target_Pop2.head()
```

## Linear plot of each z_score

```python
def make_linear_plot(z_score_data, y_lim):
    
    plt.figure(figsize=(10,5))
    for num, data in enumerate(z_score_data):
        x = []
        y = []
        for i in range(data.shape[1]):
            for j in range(data.shape[0]):
                x.append(i)
                y.append(data.iloc[j, i])

        sns.scatterplot(x = x, y = y)
        color = (0.8, 0.4, 0) if num == 0 else (0.35, 0.7, 0.9)
        fillcolor = (0.8, 0.4, 0) if num == 0 else (0.35, 0.7, 0.9)
        plt.plot([0,1,2,3,4], data.max().values, color=color, alpha=0.5)
        plt.plot([0,1,2,3,4], data.min().values, color=color, alpha=0.5, label="Pop2" if num == 0 else "C321A")
        plt.plot([0,1,2,3,4], data.mean().values, color=color)
        testx = [0,1,2,3,4,4,3,2,1,0]
        testy = [data.min().values[0],data.min().values[1],data.min().values[2],data.min().values[3],data.min().values[4],data.max().values[4],data.max().values[3],data.max().values[2],data.max().values[1],data.max().values[0]]
        plt.fill(testx,testy,color=fillcolor,alpha=0.5)
        plt.ylim(y_lim)
        plt.xticks([0,1,2,3,4], data.columns.str[0:7])
    plt.legend()
    plt.show()

# Make cluster1 (target cluster) plot.
make_linear_plot([target_Pop2, target_C321A], [-2.0, 3.0])

# Make cluster2 (control cluster) plot.
make_linear_plot([control_Pop2, control_C321A], [-3.0, 3.0])
```
